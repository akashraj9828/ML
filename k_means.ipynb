```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# from sklearn.cluster import 
import time
%matplotlib inline

time.time()
```

```
from sklearn.datasets.samples_generator import make_blobs
# color & markers array
col=['r','g','b','c','k','m','y']
markers=['s','P','*','d','x','^','h','1','.']

# generating random blobs of data points
X, Y = make_blobs(n_samples=1000, centers=4, n_features=2,cluster_std=1,shuffle=False, random_state=int(time.time()))
# print(len(X))

# plot orignal data with orignal labels 
i=0
# plt.subplot(1,2,1)
for c in X:
  plt.scatter(c[0],c[1],s=10,
#               c=col[Y[i]],
              alpha=0.5)
  i=i+1
plt.show()
```

```
# run Kmeans clustering on data
Kmean = KMeans(n_clusters=4)
Kmean.fit(X)
print(Kmean)
```

```
# plot clusters 
for i in range(len(X)):
  c=col[Kmean.labels_[i]]
  m=markers[Kmean.labels_[i]]
  plt.scatter(X[i][0], X[i][1],alpha=0.7, s =10,c=c ,marker=m)
i=0
for c in Kmean.cluster_centers_:
  plt.scatter(c[0],c[1], s=150, linewidths = 1, zorder = 100,alpha=1, c=col[-i], marker='o',)
  i=i+1

plt.show()

# from sklearn.metrics import classification_report
# print(classification_report(Y, Kmean.labels_))

```
